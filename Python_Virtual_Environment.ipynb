**Setting up virtual environment in Pycharm with using terminal**

*Pycharm terminal allows you to work with your command-line shell from inside the IDE.* 

]

```python
##Creating virtual environment:
#conda create -n my_env

##This command lists all the environments and their paths. 
##The `*` sign indicates the current active environment.
#conda env list

##Activating virtual environment
#conda activate my_env

##Deactivating virtual environment
#conda deactivate my_env

##Removing environment
##Deactivate the environment before removing##
#conda env remove -n my_env
```

**Package Management in Pycharm**

```python
##Listing installed packages
#conda list

##Installing a package -for example numpy
#conda install numpy

##Installing packages -for example scipy and pandas
#conda install scipy pandas

##Package install with pip
#pip install pandas

##Package install (certain version)
#conda install numpy=1.20.1

##Package install (certain version) with pip
#pip install pandas=1.4.2

##Package removing
#conda remove numpy

##Package upgrade
#conda upgrade numpy

##Package upgrade(all of it)
#conda update --all
```
